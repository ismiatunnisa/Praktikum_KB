# Data Split

Data splitting atau pemisahan data adalah metode membagi data menjadi dua bagian atau lebih yang membentuk subhimpunan data. Umumnya, data splitting memisahkan dua bagian, bagian pertama (data training) digunakan untuk mengevaluasi atau uji data dan data lainnya (data testing) digunakan untuk melatih model.

```python
data = dataset[['CAPITAL EXPENDITURE', 'AREA(sqkm)', 'DEPTH(m)', 'AREA(polygonal)','longitude','latitude']]
label = dataset['MAIN_USAGE']

data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=.3)
```

# Data Cleaning

Data cleaning adalah memperbarui,membetulkan, dan membuat data menjadi lebih rapi. Beberapa model machine learning tidak dapat bekerja dengan data yang masih berantakan. Kesalahan dalam data dapat berupa:

● Terdapat nilai data yang tidak valid (Null)

● Data dalam format yang salah

● Terdapat duplikat

```python
dataset3 = pd.read_csv('datasetrusak.csv')
dataset3.head()
```

```python
dataset_ns = dataset3.copy()
```

### Mengecek Nilai NULL

```python
dataset_ns.isna().sum()
```

```python
imputer_mean = SimpleImputer(strategy='mean')
imputer_median = SimpleImputer(strategy='median')
imputer_modus = SimpleImputer(strategy='most_frequent')
```

### Menghapus nilai NULL menggunakan MODUS

```python
dataset_ns['Reservior name'] = imputer_modus.fit_transform(dataset_ns[['Reservior name']])
dataset_ns['dam name'] = imputer_modus.fit_transform(dataset_ns[['dam name']])
dataset_ns['river'] = imputer_modus.fit_transform(dataset_ns[['river']])
dataset_ns['main basin'] = imputer_modus.fit_transform(dataset_ns[['main basin']])
dataset_ns['sub basin'] = imputer_modus.fit_transform(dataset_ns[['sub basin']])
dataset_ns['near city'] = imputer_modus.fit_transform(dataset_ns[['near city']])
dataset_ns['administration city'] = imputer_modus.fit_transform(dataset_ns[['administration city']])
dataset_ns['secondary country'] = imputer_modus.fit_transform(dataset_ns[['secondary country']])
dataset_ns['MAIN_USAGE'] = imputer_modus.fit_transform(dataset_ns[['MAIN_USAGE']])
dataset_ns['MULTI_DAMS,C,10'] = imputer_modus.fit_transform(dataset_ns[['MULTI_DAMS,C,10']])
dataset_ns['TIMELINE(status)'] = imputer_modus.fit_transform(dataset_ns[['TIMELINE(status)']])

dataset_ns.isna().sum()
```

## Mengecek Nilai Duplikat

```python
dataset_ns.duplicated().sum()
```

### Menghapus Nilai Duplikat

```python
dataset_ns.drop_duplicates(inplace=True)

dataset_ns.duplicated().sum()
```

# Encoding

Encoding adalah teknik yang mengubah variabel kategorikal menjadi variabel numerik, model dalam machine learning membutuhkan input yang bertipe numerik sehingga sangat penting bagi kita untuk melakukan encoding jika terdapat variabel yang bertipe kategorikal didalam data kita. Teknik encoding yang paling terkenal adalah Ordinal Encoding dan One-Hot Encoding


```python
dataset_ordinal = dataset3.copy()
dataset_one_hot = dataset3.copy()
```

```python
dataset3['COUNTRY'].unique()
```

### Ordinal Encoding

setiap nilai kategori yang unik akan diubah menjadi nilai integer.


```python
ordinal =OrdinalEncoder()

#encoding kan
dataset_ordinal['COUNTRY'] = ordinal.fit_transform(dataset_ordinal[['COUNTRY']]).astype('int64')

#lihat
dataset_ordinal.head()
```

```python
dataset_ordinal['COUNTRY'].unique()
```

### One Hot Encoding

merepresentasikan data bertipe kategori sebagai vektor biner yang bernilai integer(0 dan 1)

```python
temp_dataset = pd.get_dummies(dataset_one_hot[['COUNTRY']])

temp_dataset = pd.DataFrame(temp_dataset)

dataset_one_hot.drop('COUNTRY', axis=1, inplace=True)

dataset_one_hot = dataset_one_hot.join(temp_dataset)
dataset_one_hot.head(5)
```
