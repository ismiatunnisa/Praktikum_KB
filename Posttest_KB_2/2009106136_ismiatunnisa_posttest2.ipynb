### Import Library


```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('dams and reservoirs.csv')
```

### Tipe Dataset


```python
dataset.dtypes
```

### Melihat Dataset

```python
dataset
```

### Menampilkan Isi Dictionary dengan tipe objek

```python
jumlahrecord = len(dataset)
namaatr = dataset.columns
jumlahatr = len(dataset.select_dtypes(include=['object']).columns)

dictobjt = {
    'jumlah record': jumlahrecord,
    'nama attribut' : namaatr,
    'jumlah attribut' : jumlahatr,
}
pd.Series(dictobjt)
```

### Menampilkan Isi Dictionary dengan tipe angka

```python
jumlahatr = len(dataset.select_dtypes(include=['int64']).columns)
jumlahNULL = len(dataset.columns[dataset.isna().sum() > 0])
dictint = {
    'jumlah attribut' : jumlahatr,
    'jumlah attribut NULL' : jumlahNULL,
}
pd.Series(dictint)
```
